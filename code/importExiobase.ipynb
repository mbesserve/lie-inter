Import the pymrio library that will handle the download and preprocessing of Input-Output data from Exiobase. 

```python
import pymrio
```

Download the data in "product x product" format for selected years

```python
exio3_folder = "localdata/exiobase"
exio_meta = pymrio.download_exiobase3(
    storage_folder=exio3_folder, system="pxp", 
    years=range(1995,2018)
)
```

Convert and save the data in pkl format
